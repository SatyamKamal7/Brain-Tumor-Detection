```python
from tensorflow.keras.models import load_model

# Load the uploaded model
model = load_model('/content/drive/MyDrive/mainmodel/main_brain_tumor_model2.h5')

print("Model loaded successfully!")

```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the input image
img = cv2.imread('/content/drive/MyDrive/dataset/Training/glioma_tumor/gg (1).jpg')  # Replace with the image path
img = cv2.resize(img, (150, 150))  # Resize the image to the input size expected by the model
img_array = np.array(img)
img_array = img_array.reshape(1, 150, 150, 3)  # Reshape it for the model

# Display input image
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Input Image')
plt.show()
labels = ['glioma_tumor', 'meningioma_tumor', 'no_tumor', 'pituitary_tumor']

# Predict the class of the input image
predicted_class = model.predict(img_array)[0]
predicted_label = labels[np.argmax(predicted_class)]  # Replace `labels` with your list of class names
accuracy = np.max(predicted_class) * 100

# Output the prediction
print(f"The predicted class of the input image is: {predicted_label}")
print(f"The accuracy of prediction is: {accuracy:.2f}%")
```
